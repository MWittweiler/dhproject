```python
from typing import List

stopwords = [
    "domini", "sua", "atque", "haec", "siue", "enim", "erat", "ante", "sed", "post", "quia",
    "ipse", "pro", "super", "per", "ut", "ubi", "dicere", "deo", "tibi", "christo", "tua", "non",
    "omnes", "terra", "mihi", "eum", "sibi", "de", "hic", "si", "ergo", "dicitur", "iuxta",
    "propter", "est", "tamen", "sicut", "tu", "se", "uos", "christi", "tantum", "esse", "quo",
    "qui", "quasi", "tum", "ac", "nec", "contra", "unde", "qua", "quid", "in", "dominus", "ad",
    "tam", "eorum", "ille", "omnis", "semper", "aut", "ne", "nisi", "quem", "sit", "illa", "quos",
    "id", "omnia", "domino", "dum", "hoc", "deum", "usque", "quod", "eius", "quando", "eos",
    "uero", "nihil", "sunt", "sic", "quibus", "quomodo", "inter", "dei", "a", "ego", "quae", "mea",
    "quoque", "his", "potest", "sub", "sine", "ex", "nunc", "cui", "nos", "quam", "illud", "et",
    "cum", "quidem", "autem", "nobis", "ipsa", "iam", "te", "magis", "etiam", "ea", "eo", "ab",
    "uel", "neque", "me", "quis", "sum", "deus", "apud", "ita", "secundum", "nam", "aliud", "ecce",
    "pater", "unum", "primum", "tunc", "apostolus", "dicit", "at", "suo", "fuit", "es", "ait",
    "terram", "multa", "tuum", "ipsum", "illius", "meum", "uerum", "scriptum", "anima", "illi",
    "sint", "inquit", "corpus", "igitur", "dominum", "cuius", "omnibus", "habet", "ei", "quorum",
    "illis", "die", "graeco", "hinc", "homo", "solum", "tuae", "quas", "uobis", "prima", "corpore",
    "hac", "mei", "alii", "una", "sui", "loco", "tempore", "aeneas", "cur", "immo", "erit",
    "spiritus", "eodem", "christum", "nomine", "filium", "loquitur", "hominum", "christus",
    "ecclesiae", "hanc", "septuaginta", "arma", "tuis", "tuo", "eis", "e", "manu", "suum", "suis",
    "numquam", "caelo", "uerba", "sumus", "o", "an", "domus", "huius", "nomen", "habere",
    "aliquid", "nostra", "eam", "illum", "licet", "israhel", "tuam", "omnium", "suam", "facere",
    "ibi", "filius", "caput", "absque", "fuerit", "quoniam", "corpora", "manus", "ac", "per",
    "possum", "uel", "uero", "ut", "quicumque", "trans", "est", "autem", "sive", "etiam", "quare",
    "infra", "is", "post", "ne", "aut", "apud", "idem", "ab", "an", "enim", "ego", "interim",
    "neque", "in", "ubi", "quis", "quia", "haud", "sic", "suus", "quo", "quisquis", "magis",
    "quisquam", "tamen", "tam", "etsi", "quisnam", "mox", "igitur", "cum", "si", "unus", "aliquis",
    "es", "ex", "atque", "fio", "at", "nisi", "quam", "o", "inter", "quilibet", "ita", "tum",
    "sum", "necque", "quisque", "hic", "de", "ergo", "super", "modo", "ob", "aliqui", "iam",
    "quoniam", "sui", "nam", "sed", "ante", "deinde", "ad", "non", "pro", "nos", "et", "ille",
    "qui", "adhic", "quae", "sub", "nec", "ipse", "dum", "cur", "tu", "quidem", "quidlibet",
    "futuris", "possum", "uero", "alicui", "poteratis", "fuisti", "infra", "erat", "qua",
    "fuerint", "tui", "potuisses", "poteras", "fiebatis", "potuissemus", "potes", "possemus",
    "quorumcumque", "quodnam", "se", "fui", "eiusdem", "aliquos", "sibi", "potestis", "potuisset",
    "quodcumque", "fietis", "quidnam", "fieremus", "aliquarum", "nostri", "ipsorum", "hac", "quas",
    "ipsae", "fueramus", "quoquam", "potueris", "eosdem", "modo", "iam", "futuram", "nam", "fiant",
    "ii", "suam", "fuero", "eius", "pro", "possis", "fieretis", "qui", "essent", "futuros",
    "aliqua", "fiam", "poterunt", "erunt", "fueritis", "possent", "uni", "sit", "fuit", "quonam",
    "eisdem", "sive", "etiam", "eodem", "potuisse", "fuerit", "apud", "enim", "neque", "fimus",
    "esto", "cuiusque", "sic", "fuerimus", "tecum", "cuicumque", "fuissent", "fiebant",
    "quibuscumque", "id", "possimus", "quidque", "eorum", "nostrum", "ex", "potueram", "quasnam",
    "fuisse", "o", "essem", "mihi", "sum", "estis", "ea", "suis", "has", "fuitis", "possit",
    "futurarum", "quamcumque", "quibusque", "potuerimus", "hae", "futuro", "dum", "possunt",
    "aliquas", "quidem", "ero", "simus", "poteritis", "eram", "iidem", "fuisset", "fiebam", "post",
    "ipsos", "aut", "ubi", "illam", "hoc", "hi", "eo", "huic", "fies", "illum", "mox", "illas",
    "haec", "erant", "si", "unus", "aliquo", "fieret", "quique", "quemquam", "atque", "at",
    "eadem", "quilibet", "unae", "forent", "fiebat", "unis", "quisque", "super", "quicquam",
    "fiemus", "quoscumque", "potuissent", "poteramus", "ipsius", "ante", "quascumque", "a",
    "possetis", "ad", "nos", "eum", "suorum", "cuiuslibet", "tu", "unarum", "cuiuscumque",
    "poteris", "hunc", "ut", "quicumque", "est", "iis", "quaecumque", "autem", "earum", "eramus",
    "fiat", "fis", "ego", "poterit", "eris", "interim", "fuerant", "poterant", "ipsa", "futurus",
    "eundem", "haud", "suarum", "potuerunt", "suus", "posset", "illarum", "potuerim", "esset",
    "alicuius", "ei", "quacumque", "quidquam", "aliquae", "poteram", "suos", "es", "eorundem",
    "quamlibet", "aliquem", "sint", "quocumque", "possumus", "isdem", "potuissem", "quemlibet",
    "potuissets", "hic", "ergo", "fuissetis", "quamnam", "fore", "quoniam", "aliquod", "harum",
    "quem", "cuiusnam", "quemque", "illae", "easdem", "ac", "uel", "quosnam", "quarumcumque",
    "fueram", "trans", "eos", "quod", "possim", "sua", "esse", "quibusnam", "ipsi", "illius",
    "quorum", "fuerunt", "fias", "quo", "eandem", "quarumlibet", "cuius", "fi", "potuerant",
    "potero", "aliquis", "fuerim", "futurorum", "nisi", "illo", "potuerint", "quos", "aliquoius",
    "sui", "aliquoi", "fuerat", "ipsam", "essemus", "quidquid", "hos", "nec", "fueras", "ipse",
    "cuique", "foret", "quoque", "eam", "unos", "forem", "eratis", "quaelibet", "eras",
    "quemcumque", "potuisti", "cui", "erimus", "mecum", "fuissem", "quid", "quisquam", "tamen",
    "potueras", "etsi", "unum", "quisnam", "potuimus", "ipsarum", "igitur", "potuerit",
    "quorumlibet", "fio", "potuerat", "essetis", "quam", "possitis", "fueris", "mei", "futuri",
    "esses", "futura", "necque", "iisdem", "fiunt", "fient", "ob", "quarumnam", "futurum",
    "cuinam", "uno", "non", "mi", "eidem", "adhic", "quae", "sub", "aliquam", "poterimus", "sunto",
    "unorum", "horum", "fit", "ipsis", "eae", "fiebamus", "ipsas", "nobis", "potuero", "unius",
    "ab", "an", "fierent", "in", "illorum", "potuit", "quis", "quia", "nobiscum", "futurae",
    "sumus", "quisquis", "foremus", "possem", "huius", "potuitis", "tam", "posse", "fieres",
    "quarum", "possint", "his", "fuisses", "quaenam", "fitis", "inter", "poterat", "potueritis",
    "quoslibet", "tum", "ipso", "qualibet", "de", "aliqui", "una", "potest", "sed", "posses", "te",
    "fueratis", "illis", "ipsum", "quibus", "earundem", "per", "eritis", "quemnam", "sunt",
    "fiatis", "suas", "quolibet", "suum", "sis", "quare", "is", "ne", "eas", "idem", "fiamus",
    "suae", "fores", "unas", "aliquibus", "me", "eis", "hanc", "unam", "potueratis", "magis",
    "fiet", "futuras", "fiebas", "illi", "foretis", "cuiusquam", "este", "fierem", "sese", "illa",
    "cum", "sim", "potui", "illos", "suo", "cuilibet", "fuissemus", "fuimus", "ita", "sitis",
    "aliquorum", "quibuslibet", "eaedem", "cuiquam", "quorumnam", "tibi", "deinde", "quaslibet",
    "et", "ille", "illud", "erit", "fite", "cur", "e", "pedes", "postea", "nulla", "lege",
    "animam", "adhuc", "hebraeo", "spiritum", "legimus", "fidei", "alia", "patrem", "huc", "nemo",
    "suae", "apostoli", "meam", "uno", "postquam", "nequaquam", "peccatum", "quippe", "filii",
    "statim", "iterum", "potius", "modo", "circum", "armis", "omnem", "uitam", "quondam", "terrae",
    "posse", "ideo", "oculos", "fidem", "hominis", "euangelio", "sequitur", "homines", "omni",
    "dauid", "gratia", "prius", "nostri", "corporis", "mundi", "multis", "uiri", "sermone", "ipsi",
    "uirgo", "sermo", "dicitis", "aliis", "plus", "nostris", "meis", "mortem", "cetera", "sancti",
    "tanta", "ecclesia", "omne", "uirum", "habeat", "illo", "uoce", "supra", "uere", "spiritu",
    "urbem", "quantum", "sanguine", "animae", "eadem", "rex", "possit", "simul", "procul",
    "uerbum", "quidam", "patris", "hunc", "magna", "alius", "meus", "peccata", "oculis", "ore",
    "uitae", "uix", "uenit", "frater", "esset", "certe", "dies", "uirtutum", "tui", "nostrum",
    "uir", "haud", "ipso", "sanctum", "legis", "hierusalem", "quin", "fieri", "gloria", "rursum",
    "cumque", "tota", "itaque", "possumus", "ora", "legem", "hominem", "epistula", "quare", "meo",
    "domine", "multo", "bonum", "libros", "medio", "scripturarum", "ob", "parte", "caelum",
    "manibus", "forte", "dicens", "paulus", "meae", "alio", "saepe", "corde", "genus", "dedit",
    "dixit", "locum", "uita", "tempus", "i", "ista", "idem",
]

#Initialize text
def load_text(filename: str) -> list:
    with open(filename, 'r', encoding='utf-8') as file:
        text_list = file.readlines()
        text = []
        for line in text_list:
            line = line.split("\t")
            text.append(line)
    return text

def tokenizer(text: str) -> List[str]:
    """
    Simple tokenizer that splits text into words based on specified punctuation characters.

    Args:
        text (str): The string to tokenize.

    Returns:
        List[str]: A list of words from the input string.
    """
    word = ''
    separator = ' ?!-,.()[]:;\'\/\"'
    list_text = []
    text = text + '.'
    for k in text:
        if k not in separator:
            word = word + k
        else:
            if len(word) > 0:
                list_text.append(word)
                word = ''
    return list_text

# Count Words in Uppercase
def count_upper(text_list: list) -> dict:
    count_dict = {}
    for line in text_list:
        line_list = tokenizer(line[1])
        for token in line_list:
            if token[0].isupper() and token.lower() not in stopwords:
                count_dict[token] = count_dict.get(token, 0) + 1
    
    # Sort by count (highest first)
    sorted_counts = sorted(count_dict.items(), key=lambda x: x[1], reverse=True)
    return sorted_counts

text = load_text("../Texts/Punica.txt")
print(count_upper(text))
        



```
